```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import numpy 
import tensorflow as tf
```

```python
# seed 값 설정
seed = 3
numpy.random.seed(seed)
tf.random.set_seed(seed)

# 데이터를 불러 옵니다.
dataset = df.values
X = dataset[:,1:]
Y_obj = dataset[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state=seed)
```

```python
# 모델을 설정합니다.


early_stopping_callback = EarlyStopping(monitor='loss',patience=10)


model = Sequential()
model.add(Dense(12, input_dim=17, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy',
                 optimizer='adam',
                 metrics=['accuracy'])

# 모델을 실행합니다.
history = model.fit(X_train, Y_train, epochs=15
                    #, batch_size=10,
                    #validation_split=0.2
                    ,callbacks=[early_stopping_callback])

```

```python
history_dict = history.history
history_dict.keys()
```

```python
y_vloss = history.history['loss']
#val_loss = history.history['val_loss']

y_acc = history.history['accuracy']
#val_acc = history.history['val_accuracy']

x_len = numpy.arange(len(y_acc))
plt.plot(x_len, y_vloss, "o", c="red", markersize=3)
#plt.plot(x_len, val_loss, "o", c="gray", markersize=3)

#plt.plot(x_len, val_acc, "o", c="green", markersize=3)
plt.plot(x_len, y_acc, "o", c="blue",markersize=3)
plt.show()
```

# 모델 저장

```python
model.save("model.h5")
```
